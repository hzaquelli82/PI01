# Proceso ETL en archivo movies_datasets

Primero elimino las columnas que no se utilizarán en adelante, de acuerdo a la propuesta de trabajo.
Para ello importo los datos y desde pandas elimino las columnas

```python
import pandas as pd
import ast
```

```python
df = pd.read_csv('./datasets/movies_dataset.csv') #Cargamos los datos en un dataset
df.columns
```

```python
# Guardando columnas a eliminar en un dataframe aparte
columnas_eliminar = ['video','imdb_id','adult','original_title','poster_path','homepage']
df_trash = df[columnas_eliminar]
df_trash
```

Relizo primero esta eliminación para poder trabajar con menos datos y utilizar menos recursos de la máquina

Convierto release_date en formato fecha

```python
df['release_date'] = pd.to_datetime(df['release_date'], format='%Y-%m-%d')
df.info()
```

Creo la columna release_year

```python
df['release_year'] = df['release_date'].dt.year
```

Creo una columna de mes para la función de contar filmaciones por mes

```python
df['release_month'] = df['release_date'].dt.month
df.info()
```

```python
df.sample(5)
```

Para crear la columna return, primero cambio el tipo de dato de budget de int a float

```python

df['budget'] = df['budget'].astype(float)
df.info()
```

Luego creo la columna return, a partir del cociente entre revenue y budget, uso para calcular la columna una función lambda

```python
df['return'] = df.apply(lambda row: row['revenue'] / row['budget'] if row['budget'] != 0 else 0, axis=1)
df.head(5)
```

## Normalizar

Empiezo con el proceso de desanidar columnas. Primero, mediante una función convertimos los datos tipo texto a diccionarios y listas según se traten.

```python
def pre_desanidar (columna):
    '''Colocar como parámetro la columna de un dataframe para convertir str a dicc
    La función retornará un valor str como dicc o lista'''
    resultado = columna.apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else x)
    return resultado
```

```python
df['belongs_to_collection'] = pre_desanidar(df['belongs_to_collection'])
df['genres'] = pre_desanidar(df['genres'])
df['production_companies'] = pre_desanidar(df['production_companies'])
df['production_countries'] = pre_desanidar(df['production_countries'])
df['spoken_languages'] = pre_desanidar(df['spoken_languages'])

df.sample(5)
```

Para desanidar sigo el siguiente proceso: desanidar con la función json_normalize, y después extraer las columnas con datos útiles.

```python
from pandas import json_normalize

df_normalizado = pd.json_normalize(df['belongs_to_collection'])
df['franchise'] = df_normalizado['name']
del df_normalizado
df.head()
```

Para las columnas que tengan lista de diccionarios, ejecuto una función lambda que extrae los valores útiles y los convierte en una lista nueva. Para ordenar el código, creo una función para llevar a cabo esta tarea.

```python
def desanida_lista(df ,columna, name):
    """
    Esta función desanida una lista de diccionarios en una lista de valores, obtenidos de una 
    de las claves del diccionario.
    Y reemplaza la columna del df original
    
    Args:
        df (pandas.DataFrame): El dataframe original
        columna (str): nombre da la columna del df con listas de diccionario
        name (str): la clave de cada diccionario a conservar
    
    Returns:
        df (pandas.DataFrame): El dataframe con la columna desanida.
    """
    return df[columna].map(lambda x: [item[name].lower() for item in x])
```

```python
df['genres'] = desanida_lista(df, 'genres', 'name')
df['production_companies'] = desanida_lista(df, 'production_companies', 'name')
df['production_countries'] = desanida_lista(df, 'production_countries', 'name')
df['spoken_languages'] = desanida_lista(df, 'spoken_languages', 'iso_639_1')
df.head()
```

Elimino la columna original de 'belongs_of_collection'

```python
df_trash['belongs_to_collection'] = df['belongs_to_collection']
df.drop('belongs_to_collection', axis=1, inplace=True)
df.head()
```

```python
dfc = pd.read_csv('./datasets/credits.csv') #Carga de los datos en un dataframe
dfc.head()
```

Convierto los strings en listas de diccionarios

```python
dfc['cast'] = pre_desanidar(dfc['cast'])
dfc['crew'] = pre_desanidar(dfc['crew'])
dfc.head()
```

Exporto el dataset en formato json porque me facilita la interpretación de la info

```python
dfc['actores'] =  desanida_lista(dfc, 'cast', 'name') #Creo la columna 'actores' conteniendo una lista de los actores de cada película
dfc.head()
```

Para extraer el director utilizo una función lambda que selecciona de la columna crew el valor de la clave 'name' del diccionario cuyo valor en la clave 'job' sea 'Director'

A diferencia de la función lambda anterior, aqui usamos next para devolver el valor de 'name' y 'None' si no encuentra director. Esto para no generar una lista como resultado.

```python
dfc['director'] = dfc['crew'].map(lambda x: next((crew['name'] for crew in x if crew['job'] == 'Director'), None))
dfc.head()
```

Elimino del df las columnas que ya no son necesarias

```python
dfc.drop(['cast', 'crew'], axis=1, inplace=True)
dfc.head()
```
