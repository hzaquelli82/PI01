Primero eliminaremos las columnas que no se utilizarán en adelante, de acuerdo a la propuesta de trabajo.
Para ello importamos los datos y desde pandas eliminamos las columnas

```python
import pandas as pd
import ast
```

```python
df = pd.read_csv('./datasets/movies_dataset.csv') #Cargamos los datos en un dataset
df.columns
```

```python
# Guardando columnas a eliminar en un dataframe aparte
columnas_eliminar = ['video','imdb_id','adult','original_title','poster_path','homepage']
df_trash = df[columnas_eliminar]
df_trash
```

```python
# Eliminando columnas innecesarias del dataframe y almacenando en csv
df.drop(columnas_eliminar, axis=1, inplace=True)
df_trash.to_csv('./datasets/columns_trash.csv')
df.info()
```

```python
#Rellenar valores nulos con 0 en revenue y budget
df['revenue'] = df['revenue'].fillna(0)
df['budget'] = df['budget'].fillna(0)
df.head()
```

```python
#Para ahorrar tiempo se decidió eliminar las filas con datos nulos o atípicos en release_date
df.dropna(subset=['release_date'], inplace=True)
df.info()
df['release_date'].sort_values()
```

```python
indices_eliminar = df[(df['release_date'] == '1') | (df['release_date'] == '12') | (df['release_date'] == '22')].index
print(indices_eliminar)
df.drop(indices_eliminar, inplace=True)
df.info()
df['release_date'].sort_values()
```

```python
#Convertimos release_date en formato fecha
df['release_date'] = pd.to_datetime(df['release_date'], format='%Y-%m-%d')
df.info()
```

```python
df['release_year'] = df['release_date'].dt.year
```

```python
df.sample(5)
```

```python
#Convertir budget a float
df['budget'] = df['budget'].astype(float)
df.info()
```

```python
# Crear columna return
df['return'] = df.apply(lambda row: row['revenue'] / row['budget'] if row['budget'] != 0 else 0, axis=1)
df.head(5)
```

```python
def col_to_dicc (columna):
    '''Colocar como parámetro la columna de un dataframe para convertir str a dicc
    La función retornará un valor str como dicc'''
    resultado = columna.apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else x)
    return resultado
```

```python
df['belongs_to_collection'] = col_to_dicc(df['belongs_to_collection'])
df['genres'] = col_to_dicc(df['genres'])
df['production_companies'] = col_to_dicc(df['production_companies'])
df['production_countries'] = col_to_dicc(df['production_countries'])
df['spoken_languages'] = col_to_dicc(df['spoken_languages'])

df.sample(5)
```

```python
# Colocamos los datos en lowercase para evitar el error en la busqueda por errar la mayuscula
df['title'] = df['title'].str.lower()
```

```python
#Corregimos los valores de la columna 'popularity'
df['popularity'] = pd.to_numeric(df['popularity'])
```

```python
df.to_json('./datasets/movies.json', orient='records',lines=True,indent=4)
```

```python
df.to_parquet('./datasets/movies_datasets.parquet',engine='pyarrow')
```
